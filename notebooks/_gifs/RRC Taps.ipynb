```python
# Matplotlib
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
from matplotlib import animation

import seaborn as sns
import numpy as np
sns.set(rc={"figure.figsize": (10, 10)})

# Break from default darkgrid for better printing on paper
sns.set_style("whitegrid")

# External Includes

# Internal Includes
from rfml.ptradio import RRC
```

```python
rrc = RRC(alpha=0.35, sps=8, filter_span=8)
```

```python
taps = rrc._get_impulse_response(alpha=0.35, sps=8, filter_span=8)
```

```python
sample_num = np.arange(len(taps)) - 8*8
```

```python
fig, ax = plt.subplots(1,1, figsize=(10, 5))

ax.stem(sample_num, taps)
ax.set_title("RRC Impulse Response", fontsize=16, fontweight="bold", loc="left")
ax.set_ylabel("Amplitude")
ax.set_xlabel("Sample Number")
ax.set_xlim([-64, 64])

fig.savefig("/home/bflowers/rrc.pdf", format="pdf", transparent=True)

plt.show()
```
